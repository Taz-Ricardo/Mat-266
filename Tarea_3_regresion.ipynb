### C.1

```r
Operador <- c(1, 1, 1, 2, 2, 2, 3, 3, 3)
Velocidad_linea <- c(150, 175, 200, 150, 175, 200, 150, 175, 200)
Apariencia <- c(255, 246, 249, 260, 223, 231, 265, 247, 256)
data_C.318 <- data.frame(Operador, Velocidad_linea, Apariencia)
data_C.318
```

```r
Proletario<- factor(data_C.318$Operador)
is.factor(Proletario)
Proletario
levels(Proletario)
```

```r
modelo_dummy <- lm(Apariencia ~Proletario) #Modelo sin velocidad de linea.
summary(modelo_dummy)
```

```r
modelo_dummy
```

### C.2

Del summary deducimos que no hay ningún operador que haga una diferencia significativa. Ahora, sabemos que R nos entrega por defecto la regresión asumiendo que estamos trabajando con el primer operador. Con lo que es claro que el operador numero 1 tiene como promedio a 250 (intercepto), si se trabaja con el operador 2, se tendra que la apariencia sería 250-12=238, de la misma manera en caso del operador 3, tenemos que el promedio sería 250+6=256.

### C.3

```r
modelo_velocidad_linea <- lm(Apariencia ~Velocidad_linea)
summary(modelo_velocidad_linea)
```

El estadistico F muestra que no es relevante la velocidad en la apariencia, aún tomando un 10% de significancia.

```r
plot(Velocidad_linea, Apariencia, xlab='Velocidad de la linea', ylab='Apariencia', main='Velocidad de la linea vs Apariencia')
```

El R^{2} del modelo con variables dummy es mejor que el R^{2} del modelo de regresor con la velocidad de linea, pero el estadistico F es mejor para el modelo de regresión con la velocidad de linea, pero considerando que ambos estadisticos F son malos, es preferible el que tiene mejor R^{2}, es decir el con variables dummy.

```r
Surv <- titanic$Surv
Class <- titanic$Class
N <- titanic$N
Age <- titanic$Age
Sex <- titanic$Sex
```

```r
regresion_12.7.1 <- glm(cbind(Surv,N-Surv)~Class+Age+Sex,family=binomial())
summary(regresion_12.7.1)
```

### 12.7.2

```r
regresion_12.7.2 <- update(regresion_12.7.1,~(Class+Sex+Age)^2)
summary(regresion_12.7.2)
```

```r
drop1(regresion_12.7.2,test="Chisq")
```

```r
regresion_12.7.3 <- update(regresion_12.7.2,~Sex:Age)
summary(regresion_12.7.3)
```

```r
drop1(regresion_12.7.3,test="Chisq")
```
